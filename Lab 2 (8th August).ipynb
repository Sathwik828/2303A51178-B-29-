```python
import numpy as np
import matplotlib.pyplot as plt


monthly_temperatures = np.array([15.2, 16.4, 18.9, 22.5, 25.1, 28.3, 30.0, 29.4, 27.0, 23.5, 19.1, 16.0])
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])


average_temp = np.mean(monthly_temperatures)


hottest_index = np.argmax(monthly_temperatures)
coldest_index = np.argmin(monthly_temperatures)


normalized_temps = (monthly_temperatures - np.min(monthly_temperatures)) / (np.max(monthly_temperatures) - np.min(monthly_temperatures))


above_avg_months = months[monthly_temperatures > average_temp]


fahrenheit_temps = (monthly_temperatures * 9/5) + 32


plt.figure(figsize=(10, 5))
plt.plot(months, monthly_temperatures, marker='o', label='Celsius')
plt.plot(months, fahrenheit_temps, marker='s', linestyle='--', label='Fahrenheit')
plt.axhline(average_temp, color='red', linestyle=':', label='Average Temp (C)')
plt.title('Monthly Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


print("=== Temperature Metrics ===")
print(f"Average Temperature: {average_temp:.2f}C")
print(f"Hottest Month: {months[hottest_index]} ({monthly_temperatures[hottest_index]}C)")
print(f"Coldest Month: {months[coldest_index]} ({monthly_temperatures[coldest_index]}C)")
print("\nNormalized Temperatures:")
print(np.round(normalized_temps, 2))
print("\nMonths Above Average Temperature:")
print(above_avg_months)
print("\nTemperatures in Fahrenheit:")
print(np.round(fahrenheit_temps, 2))

```

```python
import pandas as pd


data = {
    'Order ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Product': ['Laptop', 'Phone', 'Headphones', 'Monitor', 'Laptop', 'Phone', 'Keyboard', 'Mouse'],
    'Quantity': [1, 3, 5, 2, 1, 4, 6, 10],
    'Price': [75000, 25000, 2000, 15000, 80000, 27000, 1500, 500]
}

df = pd.DataFrame(data)


df['Total Sale'] = df['Quantity'] * df['Price']


high_value_orders = df[df['Total Sale'] > 1000]


most_sold = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)


df['Discount'] = df['Total Sale'].apply(lambda x: 0.10 * x if x > 50000 else 0.05 * x)


df.to_excel("sales_summary.xlsx", index=False)

# Print intermediate and final results
print("=== Original DataFrame ===")
print(df)

print("\n=== High Value Orders (Total Sale > 1000) ===")
print(high_value_orders)

print("\n=== Most Sold Products ===")
print(most_sold)

print("\n Sales summary has been saved to 'sales_summary.xlsx'")
```
