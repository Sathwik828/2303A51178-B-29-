```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/traffic_source.csv")  # Columns: Source, Visits

# Plot bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(data["Source"], data["Visits"], color="skyblue", edgecolor="black")

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2,
             str(height), ha="center", va="bottom", fontsize=10, fontweight="bold")

# Labels and title
plt.xlabel("Traffic Source")
plt.ylabel("Number of Visits")
plt.title("Website Traffic by Source", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/monthly_rainfall.csv")  # Columns: Month, Rainfall

# Define seasonal colors
season_colors = {
    "Winter": "lightblue",   # Dec, Jan, Feb
    "Summer": "orange",      # Mar, Apr, May
    "Monsoon": "green",      # Jun, Jul, Aug, Sep
    "Autumn": "brown"        # Oct, Nov
}

# Map months to seasons
season_map = {
    "January": "Winter", "February": "Winter", "December": "Winter",
    "March": "Summer", "April": "Summer", "May": "Summer",
    "June": "Monsoon", "July": "Monsoon", "August": "Monsoon", "September": "Monsoon",
    "October": "Autumn", "November": "Autumn"
}

# Assign colors based on season
colors = [season_colors[season_map[month]] for month in data["Month"]]

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(data["Month"], data["Rainfall"], color=colors, edgecolor="black")

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2,
             str(height), ha="center", va="bottom", fontsize=9, fontweight="bold")

# Labels and title
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Average Monthly Rainfall", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
