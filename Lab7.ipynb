```python
import pandas as pd
import matplotlib.pyplot as plt

# Load global CO2 emissions data
df = pd.read_csv('/content/contentco2emissions.csv')

# Plot CO2 emissions over time
plt.plot(df['Year'], df['CO2Emissions'], marker='o', linestyle='-', color='b')
plt.title('Global CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions in million metric tons')
plt.grid(True)
plt.show()

# Load student scores data
df = pd.read_csv('/content/contentstudentscores.csv')

# Plot student performance over exams
plt.plot(df['ExamNumber'], df['Score'], marker='o', linestyle='-', color='g')
plt.title('Student Performance Over Exams')
plt.xlabel('Exam Number')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```
